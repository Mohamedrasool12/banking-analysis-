```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('/Users/dam115/Downloads/final.csv')
data.count()


```

```python

grouped = data.groupby('N_CUSTOMER_ID')

results = {}


for customer, group in grouped:
    
    unique_accounts = group['ACCOUNT_NUMBER'].unique()
    
    total_accounts = len(group['ACCOUNT_NUMBER'].unique())
    
    # Check if all bank accounts have been used
    if len(unique_accounts) == total_accounts:
        results[customer] = "All bank accounts used"
    else:
        results[customer] = "Some bank accounts not used"

# Output the results
for customer, result in results.items():
    print(f"Customer {customer}: {result}")
```

```python


# Group the data by customers
grouped = data.groupby('N_CUSTOMER_ID')

# List to store customers who used all account numbers
used_all_accounts_customers = []

# Iterate over each customer group
for customer, group in grouped:
    # Extract unique bank accounts used by the customer
    unique_accounts = group['ACCOUNT_NUMBER'].unique()
    
    # Total number of bank accounts the customer has
    total_accounts = len(group['ACCOUNT_NUMBER'].unique())
    
    # Check if all bank accounts have been used
    if len(unique_accounts) == total_accounts:
        used_all_accounts_customers.append(customer)

customer
```
